**Question 1: What is a random variable in probability theory?**

Answer: A random variable is a function that assigns a numerical value to each outcome in the sample space of a random experiment. It maps outcomes of uncertain processes to real numbers so we can analyze probabilities, expectations, variances, etc. Random variables are usually denoted by capital letters.

**Question 2: What are the types of random variables?**

Answer: Two main types:-

  - **Discrete random variables:** take a countable set of values (e.g., roll of a die). They are described by a probability mass function (PMF).

  - **Continuous random variables:** take values in a continuous range (e.g., heights). They are described by a probability density function (PDF).


**Question 3: Explain the difference between discrete and continuous distributions.**

Answer:

  - **Discrete distribution:** probability concentrated on individual points.P(X=x) is positive for each possible point x, and probabilities sum to 1.
   Examples: Binomial, Poisson.

   - **Continuous distribution:** probability is spread over an interval; the probability of any exact point is 0. A PDF
f(x) is used and probabilities are obtained by integration:
P(a≤X≤b)=∫ f(x) dx
 Examples: Normal, Exponential.

**Question 4: What is a binomial distribution, and how is it used in probability?**

Answer: The binomial distribution models the number of successes in n independent Bernoulli trials (each with success probability
p). The PMF is P(X=k) = (n k) p^k × (1-p)^ (n-k), k=0,1,...,n.


Used to model counts of successes (e.g., number of defective items in a sample, number of heads when flipping a coin
n times).

**Question 5: What is the standard normal distribution, and why is it important?**

Answer: The standard normal distribution is the normal (Gaussian) distribution with mean = 0 and standard deviation = 1; its PDF is ⌀(z) = 1 / ✓2₶ e^(-z²/2)

It's important because any normal distribution can be standardized to the normal by Z=(X−μ)/σ, which lets us use standard normal tables or functions for probabilities. It underpins many inference procedures and the Central Limit Theorem.

**Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

Answer: CLT (informal): for a large sample size

n, the sampling distribution of the sample mean Xˉ of i.i.d. observations with finite mean μ and variance σ² approaches a normal distribution:

Xˉ ≈ N(μ,σ²/n )

regardless of the original population distribution. It's critical because it justifies using normal-based inference (confidence intervals, hypothesis testing) for sample means even when the population distribution is unknown.

**Question 7: What is the significance of confidence intervals in statistical analysis?**

Answer: A confidence interval gives a range of plausible values for an unknown population parameter (e.g., a mean) computed from sample data. A 95% confidence interval constructed by a correct method will contain the true parameter in 95% of repeated samples. It quantifies uncertainty around point estimates and helps in decision-making.

**Question 8: What is the concept of expected value in a probability distribution?**

Answer: The expected value (mean) of a random variable is its long-run average value. For a discrete variable X with PMF p(x), E[X] = ∑ xp(x). For a continuous variable with PDF f(x), E[X] = ∫ xf(x) dx. It summarizes the central tendency of the distribution.



**Question 9: Write a Python program to generate 1000 random numbers from a normal**
**distribution with mean = 50 and standard deviation = 5. Compute its mean and standard**
**deviation using NumPy, and draw a histogram to visualize the distribution.**
**(Include your Python code and output in the code box below.)**

Answer:



```python

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(50)

data = np.random.normal(loc=50, scale=5, size=1000)

mean = data.mean()

std_pop = data.std(ddof=0)
std_sample = data.std(ddof=1)

print("Mean :", mean)
print("Standard Deviation:", std_pop)
print("Sample:", std_sample)

plt.hist(data, bins=30)
plt.title("Histogram of 1000 samples")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.**
**daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]**

**● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.**

**● Write the Python code to compute the mean sales and its confidence interval.**
**(Include your Python code and output in the code box below.)**

Answer:
    - Treat the 20 daily sales as a sample of size n=20. By the CLT, the sampling distribution of the sample mean Xˉ is approximately normal with mean
μ and standard error σ/✓n for sufficiently large n. Because
n=20 is modest, use the t-distribution to account for extra uncertainty from estimating σ with the sample standard deviation s.

- Compute the sample mean xˉ and sample standard deviation s. The standard error is SE = s/n. The 95% CI is

xˉ ± t0.975,df=n-1 x SE

where t0.975,df is the 97.5th percentile of a t-distribution with df=n-1

```python
import numpy as np
from scipy import stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
n = len(daily_sales)
mean_sales = np.mean(daily_sales)
std = np.std(daily_sales, ddof=1)

se = std / np.sqrt(n)

alpha = 0.05
df = n-1

t_crit = stats.t.ppf(1-alpha/2, df)

ci_low = mean_sales - t_crit*se
ci_upp = mean_sales + t_crit*se

print("n=", n)
print("Mean Sales:", mean_sales)
print("Standard Error:", se)
print("Standard Deviation:", std)
print("T_critical:", t_crit)
print("95% CI {:.6f}, {:.6f}".format(ci_low,ci_upp) )
```
